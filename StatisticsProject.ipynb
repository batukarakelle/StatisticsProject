```python
import pandas as pd
file_path = "C:\DATASHEET\Dataset Assignment.xlsx"
df = pd.read_excel(file_path)
print(df)
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)

print("Mean values:")
print(mean_values)
print("\nMedian values:")
print(median_values)
```

```python
summary_stats = pd.DataFrame({'Mean': mean_values, 'Median': median_values})
```

```python
summary_stats.to_excel('summary_statistics.xlsx', index=True)
```
